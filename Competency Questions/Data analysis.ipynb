## QC1 Qual a porcentagem de mulheres que ingressaram em cursos de STEM no Brasil por ano?

```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
df = pd.read_csv("QC1.csv")
df
```

```python
print("Cerca de 2 a 3 mulheres ingressaram a cada 10 pessoas, com a média mais próxima de",df['qtde_mulheres'].mean(),"mulheres.")
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['year'], df['fresher_female_total'], label='Mulheres ingressantes', color='indianred')
plt.bar(df['year'], df['fresher_male_total'], label='Homens ingressantes', bottom=df['fresher_female_total'], color='sandybrown')
plt.ylabel('Quantidade de pessoas ingressantes')
plt.title('Quantidade de pessoas ingressantes em cursos de STEM no Brasil por ano')
plt.xticks(df['year'])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['percentual_mulheres'],color='purple') 
plt.xlabel("Ano")  
plt.ylabel("Percentual de mulheres")  
plt.title("Percentual de mulheres que ingressaram em cursos de STEM no Brasil por ano")  
plt.xticks(df['year'])
plt.show() 
```

## QC2 Qual a porcentagem de mulheres matriculadas nos cursos de Sistemas de Informação por estados brasileiros de 2019 a 2022?

```python
df = pd.read_csv("QC2.csv")
df
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df['state_name'], df['enrolled_female2022_total'], color='lightblue', label='2022') 
plt.plot(df['state_name'], df['enrolled_female2021_total'],color='indianred', label='2021') 
plt.plot(df['state_name'], df['enrolled_female2020_total'],color='purple', label='2020') 
plt.plot(df['state_name'], df['enrolled_female2019_total'],color='sandybrown', label='2019') 
plt.legend(loc='best')
plt.xlabel("Estados brasileiros")  
plt.ylabel("Percentual de mulheres")  
plt.title("Percentual de mulheres atriculadas nos cursos de Sistemas de Informação por estados brasileiros de 2019 a 2022")  
plt.xticks(df['state_name'],rotation=45, ha="right")

plt.show() 
```

## SBC1 Evolução das Matrículas de Estudantes por Gênero entre 2009 e 2022 por grandes áreas STEM

```python
df = pd.read_csv("SBC1.csv")
df
```

```python
df_tic = df[df["general_name"] == "Computacao E Tecnologias Da Informacao E Comunicacao (Tic)"]
plt.figure(figsize=(10, 6))
plt.bar(df_tic['year']- 0.125, df_tic['enrolled_female_total'], label='Mulheres', width=0.25, color='indianred')
plt.bar(df_tic['year']+ 0.125, df_tic['enrolled_male_total'], label='Homens', width=0.25, color='sandybrown')
plt.ylabel('Quantidade de pessoas matriculadas')
plt.title('Evolução das Matrículas de Estudantes por Gênero entre 2009 e 2022 em Tecnologia')
plt.xticks(df_tic['year'])
plt.legend()
plt.show()
```

```python
df_engenharia = df[df["general_name"] == "Engenharia, Producao E Construcao"]
plt.figure(figsize=(10, 6))
plt.bar(df_engenharia['year']- 0.125, df_engenharia['enrolled_female_total'], label='Mulheres', width=0.25, color='indianred')
plt.bar(df_engenharia['year']+ 0.125, df_engenharia['enrolled_male_total'], label='Homens', width=0.25, color='sandybrown')
plt.ylabel('Quantidade de pessoas matriculadas')
plt.title('Evolução das Matrículas de Estudantes por Gênero entre 2009 e 2022 em Engenharia')
plt.xticks(df_engenharia['year'])
plt.legend()
plt.show()
```

```python
df_ciencias = df[df["general_name"] == "Ciencias Naturais, Matematica E Estatistica"]
plt.figure(figsize=(10, 6))
plt.bar(df_ciencias['year']- 0.125, df_ciencias['enrolled_female_total'], label='Mulheres', width=0.25, color='indianred')
plt.bar(df_ciencias['year']+ 0.125, df_ciencias['enrolled_male_total'], label='Homens', width=0.25, color='sandybrown')
plt.ylabel('Quantidade de pessoas matriculadas')
plt.title('Evolução das Matrículas de Estudantes por Gênero entre 2009 e 2022 em Ciências e Matemática')
plt.xticks(df_ciencias['year'])
plt.legend()
plt.show()
```
